```python
import pandas as pd
from datasets import Dataset
from tqdm import tqdm

csv_path = "Successful Test Flights/combined_dataset_successful.csv"

# Adjust this to fit your RAM
chunksize = 100_000
datasets_list = []

# Count total rows (for progress bar length)
total_rows = sum(1 for _ in open(csv_path)) - 1  # subtract header
total_chunks = total_rows // chunksize + 1

for chunk in tqdm(pd.read_csv(csv_path, dtype=str, chunksize=chunksize),
                  total=total_chunks, desc="Processing chunks"):
    ds_chunk = Dataset.from_pandas(chunk, preserve_index=False)
    datasets_list.append(ds_chunk)
```

```python

from datasets import concatenate_datasets

# Concatenate all chunks
full_dataset = concatenate_datasets(datasets_list)
```

```python
full_dataset.push_to_hub("alexducrest/successful_flight_dataset")
```
